# Descripción del dataset: `heart.csv`

Este dataset esta enfocado en enfermedades cardiacas y contiene información médica de pacientes, se utilizará para realizar un análisis exploratorio de datos (EDA) que ayudará a entender las características más importantes, detectar posibles patrones y relaciones entre variables, así como preparar los datos para la construcción de modelos predictivos.

### Características del dataset:
- **age**: Edad del paciente en años.
- **sex**: Género del paciente (1 = hombre, 0 = mujer).
- **cp**: Tipo de dolor de pecho (1: angina típica, 2: angina atípica, 3: dolor no anginal, 4: asintomático).
- **trestbps**: Presión arterial en reposo (en mm Hg).
- **chol**: Nivel de colesterol sérico (en mg/dl).
- **fbs**: Nivel de azúcar en sangre en ayunas (> 120 mg/dl, 1 = verdadero, 0 = falso).
- **restecg**: Resultados del electrocardiograma en reposo (0: normal, 1: con anomalía, 2: hipertrofia ventricular probable o definitiva).
- **thalach**: Frecuencia cardíaca máxima alcanzada.
- **exang**: Angina inducida por ejercicio (1 = sí, 0 = no).
- **oldpeak**: Depresión del ST inducida por el ejercicio en relación con el reposo.
- **slope**: Pendiente del segmento ST del ejercicio pico (1: pendiente ascendente, 2: plana, 3: descendente).
- **ca**: Número de vasos principales coloreados por fluoroscopía (0-3).
- **thal**: Resultado de la prueba de talio (3: normal, 6: defecto fijo, 7: defecto reversible).
- **target**: Variable objetivo (1 = presencia de enfermedad cardíaca, 0 = ausencia de enfermedad cardíaca).
