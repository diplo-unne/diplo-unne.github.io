**Instrucciones**:
- Utiliza la clase `Ridge` de `sklearn` para implementar la regularización L2.
- Compara los coeficientes del modelo sin regularización con el modelo regularizado.



```python
# Solución del Ejercicio 2

from sklearn.linear_model import Ridge

modelo= Ridge(alpha=0.0)
modelo.fit(X, y)

# TODO: Ajustar un modelo de regresión con regularización L2
modelo_ridge = Ridge(alpha=1.0)
modelo_ridge.fit(X, y)

# TODO: Compara los coeficientes del modelo con y sin regularización
print("Coeficientes sin regularización: ", modelo.coef_)
print("Coeficientes con regularización L2: ", modelo_ridge.coef_)

# Realiza predicciones con el modelo regularizado
y_pred_ridge = modelo_ridge.predict(X)

# Grafica los resultados
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, y_pred_ridge, color='green', label='Predicción Ridge')
plt.title('Regresión Lineal con Regularización L2')
plt.legend()
plt.show()
```

```python
# Ajusta una curva exponencial usando máxima verosimilitud
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# TODO: Generar datos reales simulados con una relación exponencial
X = np.linspace(1, 10, 100)
y = 2 * np.exp(-0.5 * X) + np.random.normal(0, 0.1, 100)

# Definir la función de verosimilitud negativa
# TODO: Definir la función de verosimilitud para una curva exponencial
def verosimilitud_exp(params, X, y):
    A, B = params
    y_pred = A * np.exp(-B * X)
    resid = y - y_pred
    return np.sum(resid ** 2)

# Ajustar los parámetros usando scipy.optimize
# TODO: Usar minimize para encontrar los mejores parámetros
resultado = minimize(verosimilitud_exp, [1, 0.1], args=(X, y))
A_opt, B_opt = resultado.x

# Graficar los datos y el ajuste
# TODO: Graficar los datos reales y el ajuste de la curva exponencial
y_pred_opt = A_opt * np.exp(-B_opt * X)

plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, y_pred_opt, color='red', label='Ajuste Máxima Verosimilitud')
plt.title('Ajuste Exponencial con Máxima Verosimilitud')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
